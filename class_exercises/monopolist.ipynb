```python
import numpy as np
```

```python
def newton(df, d2f, x_0, max_iter=1000, tol=10**-9, eps=10**-6):

    x_last = np.inf
    x_current = x_0
    n_iter = 0

    while np.abs(x_current - x_last).all() > tol:

        if n_iter >= max_iter:
            print("Did not converge to tolerance")
            break

        x_current, x_last = x_current - df(x_current) / (d2f(x_current) - eps), x_current
        n_iter += 1
    
    return x_current, n_iter
```

```python
def gd(df, x_0, learning_rate, max_iter=1000, tol=10**-9):

    x_last = np.inf
    x_current = x_0
    n_iter = 0

    while np.abs(x_current - x_last).all() > tol:

        print(f"{n_iter=}")
        print(f"{x_current=}")
        print(f"{df(x_current)=}")

        if n_iter >= max_iter:
            print("Did not converge to tolerance")
            break

        x_current, x_last = x_current + learning_rate * df(x_current), x_current
        n_iter += 1
    
    return x_current, n_iter
```

```python
df = lambda q: 90 - 3 * q ** 2
d2f = lambda q: -6 * q
```

```python
newton(
    df=df,
    d2f=d2f,
    x_0=0
)
```

```python
gd(
    df=df,
    x_0=0,
    learning_rate=.001
)
```
